```python
from PIL import Image
import os, glob
import numpy as np
from PIL import ImageFile
import tensorflow as tf
import matplotlib.pyplot as plt
tf.__version__
```

```python
classes = []
for paths,dirname,dir_in_file in os.walk("./data"):
    for d in dirname:
        classes.append(d)
print(classes)
num_class = len(classes)
print(num_class)
```

```python
for index,label in enumerate(classes):
    dirname = "./data/"+label
    print(dirname)
    files = glob.glob(dirname+"/*.*")
    for i,file in enumerate(files):
        print(file)
```

```python
num_testdata = 2
x_train = []
x_test  = []
t_train = []
t_test  = []
labels = []
for index,label in enumerate(classes):
    dirname = "./data/"+label
    files = glob.glob(dirname+"/*.*")
    labels.append(label)
    for i,file in enumerate(files):
        image = Image.open(file)
        image = image.convert("RGB")
        image = image.resize((28,28))
        data = np.asarray(image)
        if i < num_testdata:
            x_test.append(data)
            t_test.append(index)
        else:
            x_train.append(data)
            t_train.append(index)

x_train = np.array(x_train)
x_test = np.array(x_test)
t_train = np.array(t_train)
t_test = np.array(t_test)
print(x_train.shape)
print(t_test.shape)
```
